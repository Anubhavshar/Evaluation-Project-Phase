# Scalling of Data

```python
x=df.drop(['W'],axis=1)
y=df['W']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

# VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif =pd.DataFrame()
vif['vif'] = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
vif['Features'] = x.columns
vif
```

```python
vif = pd.DataFrame()
vif["vif"]= [ variance_inflation_factor(x_scaled, i)
               for i in range ( x_scaled.shape[1])]
vif['features']= x.columns
vif 
```

Observation:

1.We can see that most of the components are correlated with the target variable ,some are moderately or some are highly
correlated .
2.The reason behind it,is scalling of data,that we have not done till now
3.We can remove this by first scalling the data then using PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
x_pca = pca.fit_transform(x_scaled)
plt.figure(figsize=(6,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.xlabel('Number of Components')
plt.ylabel('Variance %')
plt.title('Explained variance Ratio')
plt.grid()
```

Observation:

We can see the first eight components attribute to 95% of variance so will take 8 components for prediction

```python
pca_new = PCA(n_components=8)
x_new = pca_new.fit_transform(x_scaled)
```

```python
p_x=pd.DataFrame(x_new,columns=np.arange(8))
p_x
```

```python

vif= pd.DataFrame()
vif['VIF']= [variance_inflation_factor(p_x.values,i) for i in range(p_x.shape[1])]
vif['Features']= p_x.columns
vif
```

Observation:

The problem of multicollinearity is solved

# Modelling Data

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
```

```python
from sklearn.linear_model import  Ridge
from sklearn.linear_model import  Lasso
from sklearn.ensemble import  GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(p_x,y,test_size = 0.33, random_state=43)
print('Training feature matrix size:',x_train.shape)
print('Training target vector size:',y_train.shape)
print('Test feature matrix size:',x_test.shape)
print('Test target vector size:',y_test.shape)
```

```python
maxR2_score=0
maxRS=0
for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(p_x, y, random_state=i, test_size=.25)
    lin_reg=LinearRegression()
    lin_reg.fit(x_train,y_train)
    y_pred=lin_reg.predict(x_test)
    R2=r2_score(y_test,y_pred)
    if R2>maxR2_score:
        maxR2_score=R2
        maxRS=i
print('Best R2 Score is', maxR2_score ,'on Random_state', maxRS)
```

# Linear Regression

```python
x_train,x_test,y_train,y_test = train_test_split(p_x,y,test_size = 0.25, random_state=45)
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
lin_reg.score(x_train,y_train)
y_pred=lin_reg.predict(x_test)
print('Mean absolute error :', mean_absolute_error(y_test,y_pred))
print('Mean squared error :', mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
print('The r2 score is:',r2_score(y_test,y_pred))
print('\n')
score = cross_val_score(lin_reg, p_x,y, cv = 5)

print("Score :" ,score)
print("Mean Score :",score.mean())
```
